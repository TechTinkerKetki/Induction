```python
import os
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
```

```python
model_path='/best_model.keras'
model=load_model(model_path)
print('loaded')
```

```python
import zipfile
zip_path = "/content/New_Data.zip"
extract_path = "/content/New_Data"

# Extract the ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Extraction complete!")
AI_path=os.path.join(extract_path,'AI')
Real_path=os.path.join(extract_path,'Real')
```

```python
#Image preprocessing
test_datagen = ImageDataGenerator(rescale=1.0/255)
test_generator = test_datagen.flow_from_directory(
    directory=extract_path,
    target_size=(128, 128),
    batch_size=20,
    class_mode=None,
    shuffle=False
)
image_filename=test_generator.filenames
```

```python
#training the data

predictions = []
valid_filenames = []

for i in range(len(test_generator.filenames)):
    try:
        image = test_generator[i]
        prediction = model.predict(image)
        label = "AI" if prediction < 0.5 else "Real"
        predictions.append(label)
        valid_filenames.append(test_generator.filenames[i])

    except Exception as e:
        print(f"Skipping corrupted image: {test_generator.filenames[i]}, Error: {e}")
```

```python
#csv preparation
df = pd.DataFrame({"filename": valid_filenames, "label": predictions})

# Save CSV
df.to_csv("submission13.csv", index=False)
print("CSV file saved as submission13.csv")
```
